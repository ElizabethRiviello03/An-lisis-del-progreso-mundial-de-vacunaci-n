# **Mostrar la estructura y tipo de datos de cada columna**

```python
import pandas as pd

vacunas = pd.read_csv("country_vaccinations.csv")
print(vacunas.head())

# Convertir 'date' columna to datetime objects
vacunas['date'] = pd.to_datetime(vacunas['date'], format='%d/%m/%Y')

# Filas de filtros donde el a침o de 'date" columna es menos que 2021
vacunas = vacunas[vacunas['date'].dt.year < 2021]
print('Tama침o del resultado: ', vacunas.shape)
vacunas.head()
```

# **Determinar la cantidad de vacunas aplicadas de compa침칤a**

```python
import pandas as pd

# Dividir la cadena de la columna 'vaccines' en nombre de empresas farmaceuticas de vacunas
# Despues crear una nueva fila para cada compa침칤a de vacunas del registro
#Finalmente agrupar por compa침칤a de vacunas individual y sumar'total_vaccinations'.
vaccine_company_vaccinations = vacunas.assign(
    individual_vaccine_company=vacunas['vaccines'].str.split(', ')
).explode('individual_vaccine_company').groupby('individual_vaccine_company')['total_vaccinations'].sum().sort_values(ascending=False)

print("Cantidad total de vacunas aplicadas por compa침칤a:")
display(vaccine_company_vaccinations)
```

# **Obtener la cantidad de vacunas aplicadas en todo el mundo**

```python
total_global_vaccinations = vacunas['total_vaccinations'].sum()

print(f"La sumatoria total de vacunas aplicadas a nivel global es: {total_global_vaccinations:.0f}")
```

# **Calcular el promedio de vacunas aplicadas por pa칤s**

```python
promedio_vacunas_por_pais = vacunas.groupby('country')['total_vaccinations'].mean()
print("Promedio de vacunas aplicadas por pa칤s:")
display(promedio_vacunas_por_pais)
```

# **Determinar la cantidad de vacunas aplicadas el d칤a 29/01/21 en todo el mundo**

```python
import pandas as pd

vacunas_full = pd.read_csv("country_vaccinations.csv")

# Convertir 췂date" column a fecha
vacunas_full['date'] = pd.to_datetime(vacunas_full['date'], format='%d/%m/%Y')

# Filtar por la fecha especifica 29/01/2021
specific_date = pd.to_datetime('29/01/2021', format='%d/%m/%Y')
vaccinations_on_specific_date = vacunas_full[vacunas_full['date'] == specific_date]

# Sumar el 'total_vaccinations' por el d칤a 29/01/2021 en todos los pa칤ses
total_vaccinations_29_01_2021 = vaccinations_on_specific_date['total_vaccinations'].sum()

print(f"La cantidad de vacunas aplicadas en todo el mundo el 29/01/2021 fue: {total_vaccinations_29_01_2021:.0f}")
```

# **Crear un dataframe con las diferencias de aplicaci칩n diarias**

```python
import pandas as pd

# Convertir la columna'date' a fecha
vacunas['date'] = pd.to_datetime(vacunas['date'], format='%Y-%m-%d', errors='coerce')

# Ordenar el dataframe por pa칤s y fecha
vacunas_sorted = vacunas.sort_values(by=['country', 'date'])

# Calcular la diferencia diaria de 'total_vaccinations' por cada pa칤s
daily_differences = vacunas_sorted.groupby('country')['total_vaccinations'].diff()

# Crear un nuevo dataframe que incluya esas diferencias diarias
daily_differences_df = vacunas_sorted[['country', 'date', 'total_vaccinations']].copy()
daily_differences_df['daily_increase'] = daily_differences

print("DataFrame con las diferencias de aplicaci칩n diarias:")
display(daily_differences_df.head())
```

# **Almacenar los DataFrames en un archivo de Excel**

```python
!pip install xlsxwriter
import pandas as pd

# Definir el nombre del archivo
excel_file_name = 'vacunacion_data.xlsx'

# Crear un Pandas Excel de los dataframes realizados
with pd.ExcelWriter(excel_file_name, engine='xlsxwriter') as writer:
    # Guardar el original 'vacunas_full' DataFrame
    if 'vacunas_full' in locals() and not vacunas_full.empty:
        vacunas_full.to_excel(writer, sheet_name='Vacunas_Original', index=False)

    # Guardar 'vacunas' DataFrame
    if 'vacunas' in locals() and not vacunas.empty:
        vacunas.to_excel(writer, sheet_name='Vacunas_Procesadas', index=False)

    # Guardar 'vaccine_company_vaccinations' DataFrame
    if 'vaccine_company_vaccinations' in locals() and not vaccine_company_vaccinations.empty:
        vaccine_company_vaccinations.to_frame(name='Total_Vaccinations').to_excel(writer, sheet_name='Vacunas_Por_Compania')

    # Guardar 'promedio_vacunas_por_pais' DataFrame
    if 'promedio_vacunas_por_pais' in locals() and not promedio_vacunas_por_pais.empty:
        promedio_vacunas_por_pais.to_frame(name='Promedio_Vacunas').to_excel(writer, sheet_name='Promedio_Por_Pais')

    # Guardar 'vaccinations_on_specific_date' DataFrame
    if 'vaccinations_on_specific_date' in locals() and not vaccinations_on_specific_date.empty:
        vaccinations_on_specific_date.to_excel(writer, sheet_name='Vacunas_29_Ene_2021', index=False)

    # Guardar 'daily_differences_df' DataFrame
    if 'daily_differences_df' in locals() and not daily_differences_df.empty:
        daily_differences_df.to_excel(writer, sheet_name='Diferencias_Diarias', index=False)

    # Guardar 'conCantidad' DataFrame
    if 'conCantidad' in locals() and not conCantidad.empty:
        conCantidad.to_excel(writer, sheet_name='Con_Cantidad_Vacunas', index=False)

    # Guardar 'vacunas_antes_20_dic_2020' DataFrame
    if 'vacunas_antes_20_dic_2020' in locals() and not vacunas_antes_20_dic_2020.empty:
        vacunas_antes_20_dic_2020.to_excel(writer, sheet_name='Antes_20_Dic_2020', index=False)

    # Guardar 'vacunas_pfizer' DataFrame
    if 'vacunas_pfizer' in locals() and not vacunas_pfizer.empty:
        vacunas_pfizer.to_excel(writer, sheet_name='Vacunas_Pfizer', index=False)

    # Guardar 'vacunas_por_dia' DataFrame
    if 'vacunas_por_dia' in locals() and not vacunas_por_dia.empty:
        vacunas_por_dia.to_excel(writer, sheet_name='Vacunas_Por_Dia', index=False)

print(f"Todos los DataFrames se han guardado en '{excel_file_name}'")
```
